```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def lkf(T, Y, U, mu0, Sigma0, A, B, C, Q, R):
    '''Linear Kalman Filter
    - 状態方程式
        x = A * x_ + B * u + w, w ~ N(0,Q)
    - 観測方程式
        y = C * x + v, v ~ N(0,R)
    
    Parameters
    ==========
    - T : ステップ数
    - Y : 観測列
    - U : 入力列
    - mu0 : 初期状態推定値
    - Sigma0 : 初期誤差共分散行列
    - A, B, C, Q, R : 状態方程式の係数と分散
    - C, R: 観測方程式の係数と分散
    
    Returns
    =======
    - M : 状態推定値列
    - SIG: 共分散行列の列
    '''
    Sigma=Sigma0 # 初期誤差共分散行列
    mu = mu0 # 初期状態推定値

    M = [mu] # 状態推定値列
    SIG=[Sigma] #共分散行列推定値列

    for i in range(T):
        # 予測ステップ
        mu_ = A * mu + B * U[i] # 事前状態推定値
        Sigma_ = A * Sigma * A.T + Q # 事前誤差共分散行列

        # フィルタリングステップ
        yi = Y[i+1] - C * mu_ #実際の観測値との誤差を算出
        g = Sigma_ * C.T * np.linalg.inv(C * Sigma_ * C.T + R)# カルマンゲインの計算

        ## 更新
        mu = mu_ + g * yi # カルマンゲインを用いて状態推定値の更新
        Sigma = Sigma_ - g * C * Sigma_ # カルマンゲインを用いて共分散行列の更新

        # 結果の格納
        M.append(mu)
        SIG.append(Sigma)
    return M,SIG

```

```python
# 状態方程式の定義
# x = A * x_ + B * u + w, w ~ N(0,Q)
theta=2*np.pi/10
A = np.mat([[np.cos(theta),-np.sin(theta)], [np.sin(theta),np.cos(theta)]])
B = np.mat([[1,0], [0,1]])
Q = np.mat([[1,0], [0,1]])
# 観測方程式の定義
# y = C * x + v, v ~ N(0,R)
C = np.mat([[1,0], [0,1]])
R = np.mat([[5,0], [0,5]])

# 観測データの生成
T = 20 # 観測数
x = np.mat([[10],[10]]) # 初期位置
X = [x] # 状態列
Y = [x] # 観測列
u = np.mat([[1],[0]]) # 入力（一定）
U = [u] # 入力列
for i in range(T):
    x = A * x + B * u + np.random.multivariate_normal([0, 0], Q, 1).T
    X.append(x)
    y = C * x + np.random.multivariate_normal([0, 0], R, 1).T
    Y.append(y)
    U.append(u)

# LKF
mu0 = np.mat([[10],[10]]) # 初期状態推定値
Sigma0 = np.mat([[1,0],[0,1]]) # 初期誤差共分散行列
M,Sigma = lkf(T, Y, U, mu0, Sigma0, A, B, C, Q, R)

# 描画
a, b = np.array(np.concatenate(X,axis=1))
plt.plot(a,b,'rs-')
a, b = np.array(np.concatenate(Y,axis=1))
plt.plot(a,b,'g^-')
a, b = np.array(np.concatenate(M,axis=1))
plt.plot(a,b,'bo-')
plt.legend(['true','observed','estimated'])
plt.axis('equal')
plt.show()
```

```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def lkf(T, Y, mu0, Sigma0, A, C, Q, R):
    '''Linear Kalman Filter
    
    - 状態方程式
        x = A * x_ + w, w ~ N(0,Q)
    - 観測方程式
        y = C * x + v, v ~ N(0,R)
    
    Parameters
    ==========
    - T : ステップ数
    - Y : 観測列
    - mu0 : 初期状態推定値
    - Sigma0 : 初期誤差共分散行列
    - A, B, C, Q, R : カルマンフィルタの係数 
    
    Returns
    =======
    - M : 状態推定値列
    '''
    Sigma=Sigma0 # 初期誤差共分散行列
    mu = mu0 # 初期状態推定値

    M = [mu] # 状態推定値列
    MU=[x]
    S=[]
    for i in range(T):
        # 推定
        mu_ = A * mu
        Sigma_ = A * Sigma * A.T + Q
        # 更新
        yi = Y[i+1] - C * mu_
        K = Sigma_ * C.T * np.linalg.inv(C * Sigma_ * C.T + R)
        mu = mu_ + K * yi
        Sigma = Sigma_ - K * C * Sigma_
        M.append(mu)
        S.append(Sigma)
    return M,S
```

```python
# 状態方程式
# x = A * x_  + w, w ~ N(0,Q)
A = np.mat([[1,0.1,-0.2], [0,1,0], [0,0,1]])
Q = np.mat([[1,0,0],[0,1,0], [0,0,1]])
# 観測方程式
# y = C * x + v, v ~ N(0,R)
C = np.mat([1,0,0])
R = np.mat([[1]])

# 観測のテストデータの生成
T = 20 # 観測数
x = np.mat([[150, 20, 10]]).T # 初期状態
X = [x] # 状態列
Y = [x] # 観測列
for i in range(T):
    x = A * x + np.random.multivariate_normal([0, 0, 0], Q, 1).T
    X.append(x)
    y = C * x + np.random.multivariate_normal([0], R, 1).T
    Y.append(y)

# LKF
mu0 = np.mat([[150],[20],[10]]) # 初期状態推定値
Sigma0 = np.mat([[0,0,0],[0,0,0],[0,0,0]]) # 初期誤差共分散行列
M,Sigma = lkf(T, Y, mu0, Sigma0, A, C, Q, R)

# 描画
a, b,c = np.array(np.concatenate(X,axis=1))
plt.plot(a,'rs-.')
plt.legend('')
a = np.array(np.concatenate(Y))
plt.plot(a,'g^-')
plt.legend('')
a,b,c = np.array(np.concatenate(M,axis=1))
plt.plot(a,'bo-')
plt.legend(['true','observed','estimated'])
#plt.axis('equal')
plt.show()
```
